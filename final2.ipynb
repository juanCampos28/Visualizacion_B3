## Pregunta de investigación 1

### ¿Como impacta el consumo de alcohol los fines de semana en el numero de materias reprobadas?

`failures = b0+ b1*Walc`

#### Variable dependiente
- failures

#### Variable independiente
- Walc 

```python
Index(['school', 'sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu',
       'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime', 'studytime',
       'failures', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
       'higher', 'internet', 'romantic', 'famrel', 'freetime', 'goout', 'Dalc',
       'Walc', 'health', 'absences', 'G1', 'G2', 'G3'],
      dtype='object')
```

```python
modelA = sm.OLS.from_formula('failures~G3', data = df)
resultA = modelA.fit()
resultA.summary()
```

### Observaciones:

- R-Squared: 0.020, con lo que podemos ver que no existe relación alguna.

`Failures = 0.1464 + 0.082*Walc`

```python
plt.figure(figsize = (14,8))
sns.set(font_scale = 1.2)

reg_plot = sns.regplot(x = 'G1',
            y = 'failures',
            data = df,
            marker = '+',
            color = 'red',
            scatter_kws = {'s': 30}
           )
reg_plot.set_title('Regression plot of failures vs Walc')

plt.show()
```

```python
RESIA = resultA.resid
RESIA
```

```python
reg_plot = sns.scatterplot(x = resultA.fittedvalues,
                            y = RESIA)
```
