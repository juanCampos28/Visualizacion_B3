```python
df = pd.read_excel ('datasets/Canada.xlsx',
                   sheet_name = 'Canada by Citizenship',
                   engine = 'openpyxl',
                   skiprows = range(20),
                   skipfooter = 200
                  )
df
```

```python
df.columns
```

```python
all(isinstance(column, str) for column in df.columns)
```

```python
df.columns = list(map(str, df.columns))
all(isinstance(column, str) for column in df.columns)
```

```python
df.drop([ 'Type',    'Coverage',                    'DEV',
           'DevName',        'Unnamed: 43',
       'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 46', 'Unnamed: 47',
       'Unnamed: 48', 'Unnamed: 49', 'Unnamed: 50'], inplace = True,
        axis = 1)
df.columns
```

```python
df.rename(columns = {'OdName' : 'Country',
                     'AreaName' : 'Continent',
                     'RegName': 'Region',
                    },
          inplace = True)
df.columns
```

```python
df['Total'] = df.iloc[:,5:].sum(axis = 1)
```

```python
df.set_index('Country', inplace = True)
```

```python
df.loc['Japan','Total']
```

```python
years = list(map(str, range(1980, 2014)))
years
```

```python
df.loc['Japan', years]
```

```python
df.loc['Japan', years].plot()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.style.use('ggplot')
```

```python
df.loc['Japan', years].plot()
```

```python
print(plt.style.available)
```

# me gustaria saber cuantas personas llegaron a haiti a canada desde 1980 - 2013 con una grafica de lineas

```python
df.loc['Haiti', years].plot()
plt.xlabel('X', fontsize = 16)
plt.xticks(fontsize = 16, rotation = 45)

plt.ylabel('Num de Haitianos', fontsize = 16)
plt.yticks(fontsize = 16, rotation = 45)

plt.title('Num de haiti', fontsize = 16)
plt.text(30,6500,'terremoto')
plt.show
```
